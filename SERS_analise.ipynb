# CÓDIGO DE ANÁLISE DOS DADOS:

# Importação e leitura dos dados:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o arquivo (certifique-se de fazer upload no Colab antes)
df = pd.read_csv("Renewable_Energy_Data.csv")

df.head()
```

# Informações gerais do dataset:

```python
print("Informações gerais do conjunto de dados:\n")
df.info()

print("\nEstatísticas descritivas:")
display(df.describe())

print("\nValores ausentes por coluna:")
print(df.isnull().sum())
```

# Correlação entre variáveis:

```python
import seaborn as sns

numeric_df = df.select_dtypes(include=['number'])

# Calcular correlação apenas entre variáveis numéricas
corr = numeric_df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlação entre Variáveis Climáticas e Eficiência (numéricas)")
plt.show()

```

# Relação entre temperatura e eficiência:

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="Temperature_C", y="Efficiency_Ratio", data=df, color="orange")
plt.title("Efeito da Temperatura na Eficiência")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Eficiência (%)")
plt.grid(True)
plt.show()
```

# Relação entre velocidade do vento e eficiência:

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="Wind_Speed_m_s", y="Efficiency_Ratio", data=df, color="teal")
plt.title("Efeito da Velocidade do Vento na Eficiência")
plt.xlabel("Velocidade do Vento (m/s)")
plt.ylabel("Eficiência (%)")
plt.grid(True)
plt.show()
```

# Relação entre radiação solar e eficiência:

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="Solar_Radiation_kWh_m2", y="Efficiency_Ratio", data=df, color="gold")
plt.title("Efeito da Radiação Solar na Eficiência")
plt.xlabel("Radiação Solar (kWh/m²)")
plt.ylabel("Eficiência (%)")
plt.grid(True)
plt.show()
```

# Relação entre chuva e eficiência:

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="Rainfall_mm", y="Efficiency_Ratio", data=df, color="blue")
plt.title("Efeito da Chuva na Eficiência")
plt.xlabel("Precipitação (mm)")
plt.ylabel("Eficiência (%)")
plt.grid(True)
plt.show()
```
